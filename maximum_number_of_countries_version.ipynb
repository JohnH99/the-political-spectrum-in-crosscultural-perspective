## Counting the number of missing values by country and variable

```python
ivs_selected.isna().sum()
```

```python
na_no_by_country_and_variable = ivs_selected.isna().groupby(ivs_selected['countries']).sum()
```

```python
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(na_no_by_country_and_variable)
```

```python
all_no_by_country_and_variable = ivs_selected.groupby(ivs_selected['countries']).sum()
```

```python
percent_by_country_and_variable = na_no_by_country_and_variable / all_no_by_country_and_variable
```
