## Concatenating the WVS and EVS dataframes

#### Solving the mismatching column

```python
print(wvs_selected_columns['X047_WVS'].unique())
```

```python
print(evs_selected_columns['X047_EVS'].unique())
```

```python
wvs_selected_columns.rename(columns = {'X047_WVS':'income_scale'}, inplace = True)
```

```python
evs_selected_columns.rename(columns = {'X047_EVS':'income_scale'}, inplace = True)
```

```python
for i in wvs_selected_columns.columns.values:
    print(i)
```

```python
for i in evs_selected_columns.columns.values:
    print(i)
```

```python
ivs_selected = pd.concat([wvs_selected_columns, evs_selected_columns])
```

## Counting the number of missing values by country and variable

```python
ivs_selected.isna().sum()
```

```python
na_no_by_country_and_variable = ivs_selected.isna().groupby(ivs_selected['S003']).sum()
```

```python
with pd.option_context('display.max_rows', None,
                       'display.max_columns', None,
                       'display.precision', 3,
                       ):
    print(na_no_by_country_and_variable)
```
